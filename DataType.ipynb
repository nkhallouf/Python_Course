### string formatting using format()

```python
num = 26
name = 'juli'
```

```python
# This is a template string {one} and {two} are placeholders
print('My number is: {one}, and my name is: {two}'.format(one=num,two=name))
```

```python
# using positional placeholders instead of named ones
print('My number is: {}, and my name is: {}'.format(num,name))
```

### Lists
Lists are used to store multiple items in a single variable, and created using square brackets>
List items are ordered, changeable, and allow duplicate values.

```python
[1,2,3]
```

```python
list = ["apple", "banana", "cherry", "apple", "cherry"]
print(list)
```

```python
['hi',1,[1,2]]
```

```python
my_list = ['a','b','c']
my_list.append('d')  # adds a single element to the end of an existing list
my_list
```

```python
# Note: The first item has index 0.
my_list[1]
```

```python
# This example returns the items from the beginning to index item = 3, but NOT including, "kiwi":


your_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(your_list[:4])
```

```python
# This example returns the items from "cherry" to the end:
your_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(your_list[2:])
```

```python
# To change the value of a specific item, refer to the index number:
your_list = ["apple", "banana", "cherry"]
your_list[1] = "blackcurrant"
print(your_list)
```

```python
# A nested list is a list within another list 
# Think of it like boxes inside boxes
nest = [1,2,3,[4,5,['target']]]
nest[3]          # outer list

```

```python
nest[3][2]       # inner list
```

```python
nest[3][2][0]   # the value
```

### Dictionaries

```python
d = {'key1':'item1','key2':'item2'}
d
```

```python
d['key1']
```

### Booleans
Booleans represent one of two values: True or False.


```python
print(10 > 9)
print(10 == 9)
print(10 < 9) 
```

### Tuples 
A tuple is a collection which is ordered, Allow Duplicates and unchangeable, Tuples are written with round brackets.

```python
t = (1,2,3)
t[0]
```

```python
# unchangeable
t[0] = 5 
```

### Sets
A set is a collection which is unordered, unchangeable, Duplicates Not Allowed,  and unindexed.

```python
{1,2,3}
```

```python
{1,2,3,1,2,1,2,3,3,3,3,2,2,2,1,1,2}
```
